```python
# Import the dependencies.
from pathlib import Path
import sqlite3
from sqlalchemy import create_engine, text
import numpy as np
import pandas as pd
import hvplot.pandas
from pathlib import Path
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Create a reference to the file. 
file_path = Path("../Data/diseases.sqlite")
```

```python
#Connect to and read from the database
conn = sqlite3.connect(file_path)
encoded_data = pd.read_sql_query("SELECT * FROM diseases", conn)
conn.close()
```

```python
#Display the data
encoded_data.head()
```

```python
#Create x and y variables
#Code for creating X and y values for a multiple regression
#https://www.w3schools.com/python/python_ml_multiple_regression.asp

y = pd.get_dummies(encoded_data['Disease'])
X = encoded_data.drop(columns='Disease')

#Display x
X.head()
```

```python
#Display y
y.head()
```

```python
#Create and fit the model
regression = LinearRegression()
regression.fit(X,y)
```

```python
#display regression coefficients
regression.coef_
```

```python
# Make predictions using the X set
predicted_y_values = regression.predict(X)
```

```python
#Create and display the metrics
score = regression.score(X, y, sample_weight=None)
r2 = r2_score(y, predicted_y_values)
mse = mean_squared_error(y, predicted_y_values)
rmse = np.sqrt(mse)
std = np.std(y)

# Print relevant metrics.
print(f"The score is {score}.")
print(f"The r2 is {r2}.")
print(f"The mean squared error is {mse}.")
print(f"The root mean squared error is {rmse}.")
print(f"The standard deviation is \n{std}.")
```
